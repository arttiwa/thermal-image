```python
import cv2
import numpy as np
import os
import glob

cv_img = []

for img in glob.glob("/Users/dis_trainee_02/jupyter-1.0.0/thermal/thermal/img/*.jpg"):
    n = cv2.imread(img)
    cv_img.append(n)

#print(cv_img)


# Function to save images with unique names
def save_image_with_unique_name(filename, image):
    if not os.path.exists(filename):
        cv2.imwrite(filename, image)
    else:
        file, ext = os.path.splitext(filename)
        counter = 1
        new_filename = f"{file}_({counter}){ext}"
        while os.path.exists(new_filename):
            counter += 1
            new_filename = f"{file}_({counter}){ext}"
        cv2.imwrite(new_filename, image)


def find_hot_regions(image, threshold, scale):
    if threshold >= 85
        threshold = 
    # Create a binary mask for regions with temperature over the threshold
    hot_mask = (image > threshold).astype(np.uint8)

    # Find contours in the binary mask
    contours, _ = cv2.findContours(hot_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Create a copy of the original image to mark
    marked_image = cv2.cvtColor(image.copy(), cv2.COLOR_GRAY2BGR)

    # Initialize variables to track the hottest regions and their maximum temperature
    max_avg_temp = 0.0
    hottest_regions = []
    max_temp = 0.0

    # area more than 4x4 pix
    filtered_contours = [cnt for cnt in contours if cv2.contourArea(cnt) > 8]  # 4x4 = 16

    # Draw filtered contours on the marked image with a red color
    cv2.drawContours(marked_image, filtered_contours, -1, (0, 0, 255), 2)

    # Calculate average temperature of hot regions and find the hottest regions
    for contour in filtered_contours:
        # Get the bounding rectangle for each contour
        x, y, w, h = cv2.boundingRect(contour)

        # Extract the region within the bounding rectangle
        region = image[y:y + h, x:x + w]

        # Calculate average temperature of the region
        avg_temp = np.mean(region)

        # Track the hottest regions based on average temperature
        if avg_temp > max_avg_temp:
            max_avg_temp = avg_temp
            hottest_regions = [(x, y, w, h)]
            max_temp = np.max(region)  # Calculate the maximum temperature in this region
        elif avg_temp == max_avg_temp:
            hottest_regions.append((x, y, w, h))

    # Highlight the hottest regions in the image
    for region in hottest_regions:
        x, y, w, h = region
        cv2.rectangle(marked_image, (x, y), (x + w, y + h), (255, 0, 0), 2)

    return marked_image, max_avg_temp*scale, max_temp*scale

# Load thermal image
thermal_image = cv2.imread('/Users/dis_trainee_02/jupyter-1.0.0/thermal/thermal/img/DJI_20221222104233_0037_T.jpg', cv2.IMREAD_GRAYSCALE)
O_thermal_image = cv2.imread('/Users/dis_trainee_02/jupyter-1.0.0/thermal/thermal/img/DJI_20221222104233_0037_T.jpg')

# Define temperature threshold
temperature_threshold = 122
scale = 2

# Find and mark hot regions, calculate average temperature, and find the maximum temperature
marked_image, avg_temperature, max_temperature = find_hot_regions(thermal_image, temperature_threshold, scale)

text = f'Avg temperature regions: {avg_temperature:.2f} degrees Celsius'
text2 = f'Max temperature region: {max_temperature:.2f} degrees Celsius'
cv2.putText(marked_image, text, (10, 20), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2, cv2.LINE_AA)
cv2.putText(marked_image, text2, (10, 60), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2, cv2.LINE_AA)

# Display the marked thermal image with the hottest pixel within hot regions
cv2.imshow('Marked Thermal Image', marked_image)
cv2.imshow('Original Thermal Image', O_thermal_image)

# Save images with unique names
save_image_with_unique_name('original.jpg', O_thermal_image)
save_image_with_unique_name('marked.jpg', marked_image)

print(f"Average temperature of hot regions: {avg_temperature:.2f} degrees Celsius")
print(f"Maximum temperature in hottest region: {max_temperature:.2f} degrees Celsius")

cv2.waitKey(0)
cv2.destroyAllWindows()
```
