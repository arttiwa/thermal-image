```python
import cv2
import numpy as np
import os 

def find_hot_regions(image, threshold):
    # Create a binary mask for regions with temperature over the threshold
    hot_mask = (image > threshold).astype(np.uint8)

    # Find contours in the binary mask
    contours, _ = cv2.findContours(hot_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Create a copy of the original image to mark
    marked_image = cv2.cvtColor(image.copy(), cv2.COLOR_GRAY2BGR)

    # area more than 4x4 pix
    filtered_contours = [cnt for cnt in contours if cv2.contourArea(cnt) > 4]  #  4x4  = 16

    # Draw filtered contours on the marked image with a red color
    cv2.drawContours(marked_image, filtered_contours, -1, (0, 0, 255), 2)

    return marked_image


def find_hottest_pixel(image):
    # Find the coordinates of the hottest pixel
    max_temp_coord = np.unravel_index(np.argmax(image), image.shape)

    # Get the temperature value at the hottest pixel
    max_temperature = image[max_temp_coord]

    return max_temp_coord, max_temperature


def mark_hottest_pixel(image, coord):
    # Create a copy of the original image to mark
    marked_image = cv2.cvtColor(image.copy(), cv2.COLOR_GRAY2BGR)

    # Draw a red circle marker at the hottest pixel
    cv2.circle(marked_image, (coord[1], coord[0]), 10, (0, 0, 255), thickness=2)

    return marked_image


def find_hottest_pixel_in_hot_regions(image, threshold):
    # Create a binary mask for regions with temperature over the threshold
    hot_mask = (image > threshold).astype(np.uint8)

    # Find contours in the binary mask
    contours, _ = cv2.findContours(hot_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Get the maximum temperature within hot regions and its coordinates
    max_temp = threshold  # Initialize with threshold
    max_temp_coord = None

    for contour in contours:
        # Get the bounding rectangle for each contour
        x, y, w, h = cv2.boundingRect(contour)
        
        # Extract the region within the bounding rectangle
        region = image[y:y+h, x:x+w]

        # Find the coordinates of the hottest pixel within the region
        coord_in_region = np.unravel_index(np.argmax(region), region.shape)

        # Convert coordinates relative to the entire image
        coord_in_image = (x + coord_in_region[1], y + coord_in_region[0])

        # Get the temperature value at the hottest pixel within the region
        temp = image[coord_in_image]

        # Update the hottest temperature and its coordinates if found
        if temp > max_temp:
            max_temp = temp
            max_temp_coord = coord_in_image

    return max_temp_coord, max_temp


# Load thermal image
thermal_image = cv2.imread(r'/Users/dis_trainee_02/jupyter-1.0.0/thermal/thermal/img/DJI_20221222104233_0037_T.JPG', cv2.IMREAD_GRAYSCALE)
O_thermal_image = cv2.imread(r'/Users/dis_trainee_02/jupyter-1.0.0/thermal/thermal/img/DJI_20221222104233_0037_T.JPG')

# Define temperature threshold
temperature_threshold = 150

# Find and mark hot regions
marked_image = find_hot_regions(thermal_image, temperature_threshold)

# Find the coordinates of the hottest pixel
max_temp_coord = np.unravel_index(np.argmax(thermal_image), thermal_image.shape)

# Get the temperature value at the hottest pixel
max_temperature = thermal_image[max_temp_coord]

# Add text with information about the hottest pixel
# cv2.putText(marked_image, text, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2, cv2.LINE_AA)

# Find the hottest pixel within hot regions
hottest_in_region_coord, hottest_in_region_temp = find_hottest_pixel_in_hot_regions(thermal_image, temperature_threshold)

markedH_image = None

# Mark the hottest pixel within hot regions
if hottest_in_region_coord is not None:
    markedH_image = mark_hottest_pixel(thermal_image, hottest_in_region_coord)

    # Add text with information about the hottest pixel within hot regions
    text = f'Hottest Pixel in Region: ({hottest_in_region_coord[1]}, {hottest_in_region_coord[0]})\nTemp: {hottest_in_region_temp:.2f} degrees Celsius'
    cv2.putText(markedH_image, text, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2, cv2.LINE_AA)
    cv2.putText(marked_image, text, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2, cv2.LINE_AA)

    # Display the marked thermal image with the hottest pixel within hot regions
    cv2.imshow('Marked Thermal Image (Hottest in Region)', markedH_image)


# Display the marked thermal image
cv2.imshow('Marked Thermal Image', marked_image)
cv2.imshow('Original Thermal Image', O_thermal_image)

directory = r'/Users/dis_trainee_02/jupyter-1.0.0/thermal'
os.chdir(directory) 
print(os.chdir(directory))
current_directory = os.getcwd()
print(current_directory)

cv2.imwrite('origina.jpg', O_thermal_image) 
cv2.imwrite('marke.jpg', marked_image) 
if markedH_image is not None:
    cv2.imwrite('markedH.jpg', markedH_image)
else:
    print("No markedH_image to save")
 


cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

color_ranges = [
    {'min_temp': 0, 'max_temp': 100, 'color': (255, 0, 0)},      # Dark Blue for 0-100C
    {'min_temp': 101, 'max_temp': 200, 'color': (0, 0, 255)},    # Red for 101-200C
    {'min_temp': 201, 'max_temp': 300, 'color': (0, 255, 0)},    # Green for 201-300C
    {'min_temp': 301, 'max_temp': 400, 'color': (255, 255, 0)},  # Yellow for 301-400C
    {'min_temp': 401, 'max_temp': 500, 'color': (255, 255, 255)} # White for 401-500C
    # Add more ranges/colors as needed
]

def map_temperature_to_color(temp, color_ranges):
    for color_range in color_ranges:
        if color_range['min_temp'] <= temp <= color_range['max_temp']:
            return color_range['color']
    return (0, 0, 0)  # Default color if temperature doesn't fall into any defined range

def apply_color_mapping(thermal_image, color_ranges):
    color_image = cv2.cvtColor(thermal_image, cv2.COLOR_GRAY2BGR)
    rows, cols = thermal_image.shape
    for i in range(rows):
        for j in range(cols):
            temperature = thermal_image[i, j]
            color = map_temperature_to_color(temperature, color_ranges)
            color_image[i, j] = color

    return color_image

def display_temperature_info(image, color_ranges):
    font = cv2.FONT_HERSHEY_SIMPLEX
    font_scale = 0.5
    font_color = (255, 255, 255)
    font_thickness = 1

    for color_range in color_ranges:
        min_temp, max_temp = color_range['min_temp'], color_range['max_temp']
        text = f'{min_temp}-{max_temp}C'
        cv2.putText(image, text, (10, (min_temp + max_temp) // 2), font, font_scale, font_color, font_thickness)

# Load the thermal image
thermal_image_path = '/Users/dis_trainee_02/jupyter-1.0.0/thermal/thermal/img/DJI_20221222104233_0037_T.jpg'
thermal_image = cv2.imread(thermal_image_path, cv2.IMREAD_GRAYSCALE)

# Apply color mapping and display temperature information
color_mapped_image = apply_color_mapping(thermal_image, color_ranges)
display_temperature_info(color_mapped_image, color_ranges)

cv2.imshow('Color-Mapped Thermal Image with Temperature Info', color_mapped_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
